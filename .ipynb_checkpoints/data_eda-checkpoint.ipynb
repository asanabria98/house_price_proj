```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('cleaned_data.csv')
data.head(5)
data.drop(columns = ['Unnamed: 0'], inplace = True)
```

```python
#Histograms numeric features
numeric_data = data.select_dtypes(exclude = ['object'])

for i in numeric_data.columns:
    print(i)
    plt.figure(figsize=(6,6))
    numeric_data[i].hist()
    plt.show()
```

```python
plt.figure(figsize=(10,10))
sns.boxplot(y = 'LotArea', data = data)
sns.stripplot(y = 'LotArea', color = 'black',data = data, alpha = 0.3)
plt.ylabel('Lot Area', size = 20)
plt.show()
```

```python
plt.figure(figsize=(10,10))
sns.boxplot(y = 'SalePrice', data = data)
sns.stripplot(y = 'SalePrice', color = 'black',data = data, alpha = 0.3)
plt.ylabel('Sales Price')
plt.show()
```

```python
plt.figure(figsize=(10,10))
sns.boxplot(y = 'GrLivArea', data = data)
sns.stripplot(y = 'GrLivArea', color = 'black',data = data, alpha = 0.3)
plt.ylabel('GrLivArea')
plt.show()
```

```python
corr_matrix = numeric_data.corr()

# Can be great to plot only a half matrix
mask = np.zeros_like(corr_matrix)
mask[np.triu_indices_from(mask)] = True

plt.figure(figsize = (12,12))
sns.heatmap(corr_matrix, mask = mask, square = True, annot = True, cmap = "Blues")
plt.show()
```
